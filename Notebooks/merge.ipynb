```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sys
sys.path.append('src')
import db_connection
import numpy as np

# Leer el dataset de Spotify
df = pd.read_csv("/home/ana/workshop2/Data/spotify_dataset.csv")

# Leer datos de la base de datos de los Grammys
query = "SELECT * FROM grammys"
conn = db_connection.conn()  # Obtener la conexión
dataframe_raw = pd.read_sql(sql=query, con=conn)

# Realizar el merge entre los datasets
df_merge = pd.merge(dataframe_raw, df, left_on="nominee", right_on="track_name", how="inner")

# Guardar el resultado en un archivo CSV
df_merge.to_csv("/home/ana/workshop2/Data/merge.csv", index=False)
```
